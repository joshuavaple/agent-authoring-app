```python
import requests
from dotenv import load_dotenv
import os


load_dotenv("../../.env")
DATABRICKS_TOKEN = os.environ["DATABRICKS_TOKEN"]
# base_url = "https://dbc-564fb500-5a75.cloud.databricks.com" # free edition, not supported
base_url = "https://adb-2598814964271509.9.azuredatabricks.net" # premium edition
```

```python
api_url = "/api/2.0/mlflow/experiments/list"
url = base_url + api_url
response = requests.get(
    url=url,
    headers={
        "Authorization":f"Bearer {DATABRICKS_TOKEN}"
    },
    params={
        "view_type":"ALL",
        "max_results":10,
    }
)
# response
print(response.status_code)
infos = []
for exp in response.json()["experiments"]:
    id = exp['experiment_id']
    name = exp['name']
    short_name = name.split("/")[-1]
    location = exp['artifact_location']
    info = (id, short_name, location)
    infos.append(info)
    print(info)
```
