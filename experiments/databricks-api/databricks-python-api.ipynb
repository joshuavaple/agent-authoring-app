```python
import mlflow


# install mlflow[databricks] first
# Using profile authenticated to free edition workspace:
mlflow.set_tracking_uri("databricks://joshuale-common") 
```

```python
response = mlflow.search_experiments(max_results=10)
```

```python
# the Experiment object follows the same attribute names as the response from REST API
infos = []
for exp in response:
    id = exp.experiment_id
    name = exp.name
    short_name = name.split("/")[-1]
    location = exp.artifact_location
    info = (id, short_name, location)
    infos.append(info)
    print(info)
```

```python
from google.protobuf.json_format import MessageToDict

dicts = [MessageToDict(e.to_proto(), preserving_proto_field_name=True) for e in response]
dicts
```
